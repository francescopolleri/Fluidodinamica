```python
import plotly.figure_factory as ff
import cmath
import numpy as np
from matplotlib import pyplot as plt

#CALCOLO DIRETTO CON COORDINATE POLARI

plt.rcParams["figure.figsize"] = [9, 9]
plt.rcParams["figure.autolayout"] = True

Omega=1; alpha=np.pi/6

radii = np.linspace(0, 0.1, 100)
thetas = np.linspace(0, 2*np.pi, 200)
theta, r = np.meshgrid(thetas, radii)

f = plt.figure()

F=Omega*r*np.pi/(4*alpha)

ax = f.add_subplot(polar=True)
ax.quiver(theta, r,Omega*r*np.pi/(4*alpha)*np.cos(np.pi*theta/(2*alpha)), -Omega*r*np.sin(np.pi*theta/(2*alpha)),color='b')

plt.show()
```

```python

#CALCOLO IN COORDINATE CARTESIANE ATTAVERSO COORDINATE POLARI

N=80
X=np.linspace(-10,10,N)
Y=np.linspace(-10,10,N)
x,y = np.meshgrid(X,Y)

r=np.sqrt(x**2+y**2)
theta=np.arctan(y/x)

Omega=1; alpha=np.pi/6

u_r=Omega*np.pi*r/(4*alpha)*np.cos(np.pi*theta/(2*alpha))
u_t=-Omega*r*np.sin(np.pi*theta/(2*alpha))

u_x=u_r*np.cos(theta)-u_t*np.sin(theta)
u_y=u_r*np.sin(theta)+u_t*np.cos(theta)

plt.rcParams["figure.figsize"] = [7, 6]

plt.quiver(x,y,u_x,u_y,color='r')
plt.show()
```

```python
#STREAMLINES PLOT

fig=ff.create_streamline(X,Y,u_x,u_y,arrow_scale=0.3,density=3.2)
fig.show()
```
