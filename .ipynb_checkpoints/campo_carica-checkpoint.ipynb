```python
import plotly.figure_factory as ff
import cmath
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=150
x=np.linspace(-10,10,N)
y=np.linspace(-15,15,N)
X,Y = np.meshgrid(x,y)
```

```python
u = X/((X**2+Y**2)**(3/2))
v = Y/((X**2+Y**2)**(3/2))
```

```python
plt.quiver(x, y, u, v, color='b')

plt.xlim(-10, 10)
plt.ylim(-15, 15)

plt.grid()
plt.show()
```

```python
fig=ff.create_streamline(x,y,u,v,arrow_scale=0.2,density=2)
fig.show()
```
