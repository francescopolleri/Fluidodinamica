```python
import plotly.figure_factory as ff
import cmath
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
pts = 100
x_range = 10
y_range = 10
colors_in_quiver = 20
headwidth=5.0
minlength=3
pivot='tail'

U=0.001
a=3
```

```python
# Initialization
x,y = np.linspace(-x_range,x_range,pts), np.linspace(-y_range,y_range, pts)
X,Y = np.meshgrid(x,y)
```

```python
# Components of H
u = U + a**2*U*(Y**2-X**2)/((X**2+Y**2)**2)
v = -2*a**2*U*X*Y/((X**2+Y**2)**2)
```

```python
plt.quiver(x, y, u, v, color='r')

plt.xlim(-10, 10)
plt.ylim(-10, 10)

plt.grid()
plt.show()
```

```python
fig=ff.create_streamline(x,y,u,v,arrow_scale=0.2,density=3)
fig.show()
```

```python
# Plot 
fig, ax = plt.subplots(figsize=(10,9))
ax.quiver(x, y, u, v,
          [pd.qcut(u.flatten(), q=colors_in_quiver, labels=False)],
          headwidth=headwidth,
          minlength=minlength,
          pivot=pivot, 
          cmap='hsv')
```
